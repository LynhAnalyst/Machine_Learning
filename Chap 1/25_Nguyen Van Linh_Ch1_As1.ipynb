```python
import pandas as pd
import numpy as np
import statistics as st
```

```python
path = 'vgsales.csv'
df = pd.read_csv(path)
df.head(10)
```

```python
# Hiển thị số dòng/cột của dữ liệu
print(df.shape)
```

```python
# Miêu tả bán hàng trên toàn thế giới
print(df['Global_Sales'].describe())
```

```python
# Thể loại game xuất hiện nhiều nhất
st.mode(df['Genre'])
```

```python
# Giá trị trung bình của EU_Sales
print(df['EU_Sales'].mean())
```

```python
# Giá trị phương sai của JP_Sales
print(df['JP_Sales'].var())
```

```python
# Độ lệch chuẩn của JP_Sales
st.stdev(df['JP_Sales'])
```

```python
# Mối tương quan giữa các cột trong dataframe
print(df.corr(method='pearson'))
```

```python
# Nhóm theo từng thể loại
gk = df.groupby('Genre')
gk.first()
```

```python
# Tổng doanh số bán hàng theo từng thể loại
print(gk['NA_Sales'].sum())
```
